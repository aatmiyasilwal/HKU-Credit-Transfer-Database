```python
import pandas as pd
```

```python
#merging and sorting the file
def merge(file_paths):
    df_list = [pd.read_csv(file) for file in file_paths]
    merged_df = pd.concat(df_list, ignore_index=True)

    merged_df['University'] = merged_df['University'].str.strip()
    merged_df['Exchange-Course-Code'] = merged_df['Exchange-Course-Code'].str.strip()
    merged_df['Exchange-Course-Title'] = merged_df['Exchange-Course-Title'].str.strip()

    merged_df['University'] = merged_df['University'].str.replace(r'^The\s+', '', regex=True)
    sorted_df = merged_df.sort_values(by='University')
    return sorted_df
```

```python
#all files in the database
file_paths = [
    'arts-credit_transfer_database.csv',
    'business-credit_transfer_database.csv',
    'compsci-credit_transfer_database.csv',
    'engg-credit_transfer_database.csv',
    'science-credit_transfer_database.csv',
    'sosci-credit_transfer_database.csv'
]  

df = merge(file_paths)
```
